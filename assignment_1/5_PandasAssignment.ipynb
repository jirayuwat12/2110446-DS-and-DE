## How many VDO's that contain at least one record (row) of "dislikes" more than "likes"?  <font color=red>DO NOT group by the data and make sure that you count only unique title!</font>

```python
vdo_id_list = set()
for idx, file_name in enumerate(['USvideos', 'GBvideos']):
  df = pd.read_csv(f'data/{file_name}.csv')
  filtered = df[df['dislikes'] > df['likes']]['video_id'].unique()
  vdo_id_list = set(filtered).union(vdo_id_list)

  print(f"for {file_name}.csv \thas {len(filtered):3} id")

print(f"for union of 2 files \thas {len(vdo_id_list):3} id")
```

## How many VDO that are trending on 22 Jan 2018 with comments more than 10,000 comments?

```python
vdo_id_list = set()
for idx, file_name in enumerate(['USvideos', 'GBvideos']):
  df = pd.read_csv(f'data/{file_name}.csv')
  filtered = df[(df['trending_date'] == '18.22.01')&(df['comment_count'] > 10000)]['video_id'].unique()
  vdo_id_list = set(filtered).union(vdo_id_list)

  print(f"for {file_name}.csv \thas {len(filtered):3} id")
print(f"for union of 2 files \thas {len(vdo_id_list):3} id")
```

## Which date that has the minimum average number of comments per VDO?

```python
date = None
min_avg = float('inf')

for idx, file_name in enumerate(['USvideos', 'GBvideos']):
  df = pd.read_csv(f'data/{file_name}.csv')
  df = df.groupby('trending_date').agg({'comment_count' : ['sum', 'count']})
  df['comment_avg'] = df['comment_count']['sum'] / df['comment_count']['count']

  min_row = df[df['comment_avg'] == df['comment_avg'].min()]
  current_date = min_row.index[0]
  current_min_avg = min_row.iloc[0, 2]
  print(f"for file {file_name}.csv \t\tminimum average from date {current_date} which is {current_min_avg:.4f}")

  if current_min_avg < min_avg:
    min_avg = current_min_avg
    date = current_date

print(f"for both 2 files \t\tminimum average from date {date} which is {min_avg:.4f}")
```
